Changing value for "Male,Female" string values to numeric values,male = 1 and female = 2

```python
def Number(str):
    if str=="male":
        return 1
    else:
        return 2
titanic["Gender"]=titanic["Sex"].apply(Number)
#we have created a new column called "gender" and
#filling it with values 1,2 based on the values of sex column
titanic.head()
```

Ques1: What is the distribution of passenger ages on the Titanic? Create a histogram to visualize the age distribution.

Histogram: Histogram is a graph showing the frequency distribution.

```python
import matplotlib.pyplot as plt


passenger_ages = titanic['Age']
plt.hist(passenger_ages, bins=10,edgecolor = 'black')

plt.xlabel('Age')
plt.ylabel('count')
plt.title('Distribution of Passenger Ages on the Titanic')

plt.show()
```

Ques2: Is there a relationship between passenger class and survival rate? plot a bar chart showing the survival rate for each passenger class.

```python
import matplotlib.pyplot as plt

survival_rate = titanic.groupby('Pclass')['Survived'].mean()
print(survival_rate)
survival_rate.plot(kind = 'bar', edgecolor = 'black')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.title('Survival Rate by Passenger Class')
plt.show()
```

Ques3: How does the fare vary across different passenger classess? create a boxplot to compare the fare distribution for each class.

```python
import matplotlib.pyplot as plt

fare_data = titanic[['Pclass', 'Fare']]

# Create a boxplot using Pandas
fare_data.boxplot(by='Pclass', column='Fare', showfliers = False)

# Customize the plot
plt.title("Fare Distribution by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
```

Ques4: What is Gender distribution on the Titanic? Visualize count of males and females using a bar chart. 

```python
import matplotlib.pyplot as plt

gender_counts = titanic['Sex'].value_counts()
gender_counts.plot(kind = 'bar', edgecolor = 'black')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Distribution on the Titanic')
plt.show()
```

Ques5: How does the survival rate differ based on the port of embarkation? Create a pie chart to display the proportion of survivors from each port.

```python
import matplotlib.pyplot as plt
import numpy as np

survivors_port = titanic.groupby('Embarked')['Survived'].sum()
total_survivors = survivors_port.sum()
proportions = survivors_port / total_survivors
plt.pie(proportions, labels = ['C', 'Q', 'S'], startangle = 90)


plt.title('Proportion of Survivors by Port of Embarkation')
plt.show()
```

Ques6: Is there a relationship between age, fare, and survival? Plot a scatter plot to examine the relationship between age, fare, and survival, using different colors for survivors and non-survivors.

```python
import matplotlib.pyplot as plt

plt.scatter(titanic[titanic['Survived'] == 1]['Age'], titanic[titanic['Survived'] == 1]['Fare'])
plt.scatter(titanic[titanic['Survived'] == 0]['Age'], titanic[titanic['Survived'] == 0]['Fare'])


plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Relationship between Age, Fare, and Survival')
plt.show()
```
